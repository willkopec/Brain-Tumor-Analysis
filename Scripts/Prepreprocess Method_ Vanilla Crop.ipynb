```python
#########################
# Load Images : Vanilla Crop
#
# Loads in all the brain scans and crops them where the mask is
# Stores the data in Preprocessed_Data/Vanilla_Crop as a pickled dictionary with keys 1, 2, 3
# and the the corresponding values are all the 224 x 224 images stored as a numpy array in a list
# 
# 1 for meningioma, 2 for glioma, 3 for pituitary tumor

import h5py
import os
import numpy as np
import pickle

import matplotlib.pyplot as plt


RAW_DATA_PATH = '/content/gdrive/MyDrive/CS 415 Project/Raw Data'
FEATURE_PATH = '/content/gdrive/MyDrive/CS 415 Project/Vanilla_Crop_Dataset.pk1'

# load_image
#
# Takes in a matfile and loads the data
# - Extracts the label
# - Extracts the tumor
#
def load_image(mat_file):
  with h5py.File(mat_file, 'r') as f:

    # Extracts the label from the data 
    label = str(int(f['cjdata']['label'][0][0]))

    # Extracts the mask
    mask = np.array(f['cjdata']['tumorMask'][:]).astype('uint8')

    # Extracts the image of the whole brain
    img = np.array(f['cjdata']['image'][:])

    # Extracts the tumor image
    img[np.where(mask == 0)] = 1 # applies the mask
    tumor = img

  return tumor, label

A = 0

if __name__ == '__main__':
  vanilla_crop = {'1':[],'2':[], '3':[]}
  for im_num in range(1, 3065):
    mat_file = os.path.join(RAW_DATA_PATH, str(im_num) + '.mat')
    img, class_num = load_image(mat_file)
    vanilla_crop[class_num].append(img)


  A = vanilla_crop
```
